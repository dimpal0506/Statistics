````python
#1.Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
#After generating the list, find the following:

#(i) Write a Python function to calculate the mean of a given list of numbers.
#Create a function to find the median of a list of numbers.

#(ii) Develop a program to compute the mode of a list of integers.

#(iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

#(iv) Write a Python function to find the geometric mean of a list of positive numbers.

#(v) Create a program to calculate the harmonic mean of a list of values.

#(vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

#(vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers.


import random
import statistics

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i) Calculate mean and median
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        return (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2
    else:
        return sorted_numbers[n//2]

# (ii) Compute mode
def calculate_mode(numbers):
    return statistics.mode(numbers)

# (iii) Calculate weighted mean
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

# (iv) Calculate geometric mean
def calculate_geometric_mean(numbers):
    product = 1
    for number in numbers:
        product *= number
    return product ** (1 / len(numbers))

# (v) Calculate harmonic mean
def calculate_harmonic_mean(numbers):
    harmonic_sum = 0
    for number in numbers:
        harmonic_sum += 1 / number
    return len(numbers) / harmonic_sum

# (vi) Calculate midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# (vii) Calculate trimmed mean
def calculate_trimmed_mean(numbers, trim_percent):
    n = len(numbers)
    trim_index = int(n * trim_percent / 100)
    sorted_numbers = sorted(numbers)
    trimmed_numbers = sorted_numbers[trim_index:-trim_index]
    return calculate_mean(trimmed_numbers)

print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean (assuming equal weights):", calculate_weighted_mean(int_list, [1] * len(int_list)))
print("Geometric Mean:", calculate_geometric_mean(int_list))
print("Harmonic Mean:", calculate_harmonic_mean(int_list))
print("Midrange:", calculate_midrange(int_list))
print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))
````

````python
#2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.
#After generating the list, find the following:

#(i) Compare the given list of visualization for the given data:
#1. Frequency & Gaussian distribution
#2. Frequency smoothened KDE plot
#3. Gaussian distribution & smoothened KDE plot

#(ii) Write a Python function to calculate the range of a given list of numbers.

#(iii) Create a program to find the variance and standard deviation of a list of numbers.

#(iv) Implement a function to compute the interquartile range (IQR) of a list of values.

#(v) Build a program to calculate the coefficient of variation for a dataset.

#(vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.

#(vii) Create a program to calculate the quartile deviation of a list of values.

#(viii) Implement a function to find the range-based coefficient of dispersion for a dataset.

import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualizations
# Convert the list to a NumPy array for easier manipulation
data = np.array(int_list2)

# 1. Frequency & Gaussian distribution
plt.figure(figsize=(12, 6))
sns.histplot(data, kde=True, stat='frequency', bins=20)
plt.title('Frequency & Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 2. Frequency smoothened KDE plot
plt.figure(figsize=(12, 6))
sns.histplot(data, kde=True, stat='frequency', bins=20, kde_kws={'bw_adjust': 1.5})
plt.title('Frequency & Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 3. Gaussian distribution & smoothened KDE plot
plt.figure(figsize=(12, 6))
sns.histplot(data, kde=True, stat='density', bins=20, kde_kws={'bw_adjust': 1.5})
plt.title('Gaussian Distribution & Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# (ii) Calculate range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Calculate variance and standard deviation
def calculate_variance_std(numbers):
    mean = sum(numbers) / len(numbers)
    variance = sum((x - mean) ** 2 for x in numbers) / len(numbers)
    std_dev = variance ** 0.5
    return variance, std_dev

# (iv) Calculate interquartile range (IQR)
def calculate_iqr(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

# (v) Calculate coefficient of variation
def calculate_coefficient_of_variation(numbers):
    mean = sum(numbers) / len(numbers)
    std_dev = np.std(numbers)
    return (std_dev / mean) * 100

# (vi) Calculate mean absolute deviation (MAD)
def calculate_mad(numbers):
    mean = sum(numbers) / len(numbers)
    mad = sum(abs(x - mean) for x in numbers) / len(numbers)
    return mad

# (vii) Calculate quartile deviation
def calculate_quartile_deviation(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return (q3 - q1) / 2

# (viii) Calculate range-based coefficient of dispersion
def calculate_range_based_coefficient_of_dispersion(numbers):
    range_val = max(numbers) - min(numbers)
    mean = sum(numbers) / len(numbers)
    return range_val / mean

print("Range:", calculate_range(int_list2))
print("Variance, Standard Deviation:", calculate_variance_std(int_list2))
print("Interquartile Range (IQR):", calculate_iqr(int_list2))
print("Coefficient of Variation:", calculate_coefficient_of_variation(int_list2))
print("Mean Absolute Deviation (MAD):", calculate_mad(int_list2))
print("Quartile Deviation:", calculate_quartile_deviation(int_list2))
print("Range-based Coefficient of Dispersion:", calculate_range_based_coefficient_of_dispersion(int_list2))
````

````python
#3.Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
class DiscreteRandomVariable:
    def __init__(self, pmf):

        self.pmf = pmf

    def expected_value(self):

        expected_value = 0
        for x, p in self.pmf.items():
            expected_value += x * p
        return expected_value

    def variance(self):
        mean = self.expected_value()
        variance = 0
        for x, p in self.pmf.items():
            variance += p * (x - mean) ** 2
        return variance

pmf = {1: 0.2, 2: 0.3, 3: 0.5}
rv = DiscreteRandomVariable(pmf)

print("Expected value:", rv.expected_value())
print("Variance:", rv.variance())
````

````python
#4.Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
import random
import statistics

def simulate_dice_rolls(num_rolls):
  rolls = []
  for _ in range(num_rolls):
    roll = random.randint(1, 6)
    rolls.append(roll)
  return rolls

def calculate_expected_value(rolls):
  return sum(rolls) / len(rolls)

def calculate_variance(rolls):
  mean = calculate_expected_value(rolls)
  variance = sum((roll - mean) ** 2 for roll in rolls) / len(rolls)
  return variance

# Simulate 10,000 rolls
num_rolls = 10000
rolls = simulate_dice_rolls(num_rolls)

# Calculate expected value and variance
expected_value = calculate_expected_value(rolls)
variance = calculate_variance(rolls)

print("Expected value:", expected_value)
print("Variance:", variance)
````

````python
#5.Create a Python function to generate random samples from a given probability distribution (e.g.binomial, Poisson) and calculate their mean and variance

import numpy as np

def generate_random_samples(distribution, params, num_samples):
  if distribution == 'binomial':
    samples = np.random.binomial(n=params[0], p=params[1], size=num_samples)
  elif distribution == 'poisson':
    samples = np.random.poisson(lam=params, size=num_samples) # Changed line: directly use params
  else:
    raise ValueError('Invalid distribution')

  return samples

def calculate_mean_and_variance(samples):
  mean = np.mean(samples)
  variance = np.var(samples)
  return mean, variance

# Example usage:
num_samples = 10000
binomial_samples = generate_random_samples('binomial', (10, 0.5), num_samples)
poisson_samples = generate_random_samples('poisson', (5), num_samples)

binomial_mean, binomial_variance = calculate_mean_and_variance(binomial_samples)
poisson_mean, poisson_variance = calculate_mean_and_variance(poisson_samples)

print("Binomial distribution:")
print("Mean:", binomial_mean)
print("Variance:", binomial_variance)

print("\nPoisson distribution:")
print("Mean:", poisson_mean)
print("Variance:", poisson_variance)
````

````python
#11.A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitorswho make a purchase). They collect data from the old and new layouts to compare.
#To generate the data use the following command:
#```python
#import numpy as np
# 50 purchases out of 1000 visitors
#old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors
#new_layout = np.array([1] * 70 + [0] * 930)
#  ```
#Apply z-test to find which layout is successful.


import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Given data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate the number of successes and sample sizes
successes = np.array([sum(old_layout), sum(new_layout)])
nobs = np.array([len(old_layout), len(new_layout)])

# Perform z-test for two proportions
z_stat, p_value = proportions_ztest(count=successes, nobs=nobs)

# Set significance level (alpha)
alpha = 0.05

# Hypothesis testing
if p_value < alpha:
    print("Reject the null hypothesis. The new layout has a significantly higher conversion rate.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the new layout is significantly better.")
````

````python
#13.A pharmaceutical company wants to determine if a new drug is effective in reducing blood pressure. They conduct a study and record blood pressure measurements before and after administering the drug.
#Use the below code to generate samples of respective arrays of blood pressure:
#```python
#before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
#after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])
#  ```
#Implement z-test to find if the drug really works or not.


import numpy as np
from scipy.stats import zscore

# Given data
before_drug = np.array([145, 150, 140, 135, 155, 160, 152, 148, 130, 138])
after_drug = np.array([130, 140, 132, 128, 145, 148, 138, 136, 125, 130])

# Calculate the difference between before and after measurements
difference = after_drug - before_drug

# Calculate the mean and standard deviation of the differences
mean_diff = np.mean(difference)
std_dev_diff = np.std(difference, ddof=1)

# Calculate the z-score
z_score = mean_diff / (std_dev_diff / np.sqrt(len(difference)))

# Set significance level (alpha)
alpha = 0.05

# Critical value for a two-tailed test at alpha = 0.05
critical_value = 1.96

# Hypothesis testing
if abs(z_score) > critical_value:
    print("Reject the null hypothesis. The drug is effective in reducing blood pressure.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the drug is effective.")
````

````python
#15.A company is testing two different website layouts to see which one leads to higher click-through rates.
#Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.
#Use the following data:
#```python
#layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
#layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

import numpy as np
from scipy.stats import ttest_ind

def ab_test(group_a, group_b):
  t_stat, p_value = ttest_ind(group_a, group_b)
  df = len(group_a) + len(group_b) - 2

  return t_stat, df, p_value

# Given data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
t_stat, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

# Set significance level (alpha)
alpha = 0.05

# Hypothesis testing
if p_value < alpha:
  print("Reject the null hypothesis. Layout B has a significantly higher click-through rate.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that one layout is significantly better than the other.")
````

````python
#17.A school district introduces an educational intervention program to improve math scores. Write a Python function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to determine if the intervention had a significant impact.
#Use the following data of test score:
#```python
#pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
#post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
#```

import numpy as np
from scipy.stats import ttest_rel

def analyze_intervention(pre_scores, post_scores):
  t_stat, p_value = ttest_rel(post_scores, pre_scores)
  df = len(pre_scores) - 1

  return t_stat, df, p_value

# Given data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform paired t-test
t_stat, df, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)

# Set significance level (alpha)
alpha = 0.05

# Hypothesis testing
if p_value < alpha:
  print("Reject the null hypothesis. The intervention had a significant positive impact on math scores.")
else:
  print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the intervention was effective.")
````

````python
#18.An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees.
#Use the below code to generate synthetic data:
#```python
# Generate synthetic salary data for male and female employees
#np.random.seed(0)  # For reproducibility
#male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
#female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
#```

import numpy as np
from scipy.stats import ttest_ind

# Generate synthetic salary data for male and female employees
np.random.seed(0)
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform independent t-test
t_stat, p_value = ttest_ind(female_salaries, male_salaries)

# Set significance level (alpha)
alpha = 0.05

# Hypothesis testing
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in average salaries between male and female employees.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude a significant salary gap.")
````
